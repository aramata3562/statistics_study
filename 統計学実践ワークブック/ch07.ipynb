## 大数の弱法則

大数の弱法則（The Weak Law of Large Numbers）とは、確率論における基本的な定理の一つで、多くの試行を行うことで、その平均が真の平均値に近づくことを示しています。具体的には、独立同分布の確率変数の平均がその期待値に収束することを意味します。

### 定理の内容
独立同分布（i.i.d.）の確率変数列 $\{X_i\}$ が存在し、それぞれの確率変数の期待値が $\mu$ であるとします。このとき、任意の正の数 $\epsilon$ に対して、次の不等式が成立します。

$$
\lim_{n \to \infty} P\left(\left| \frac{1}{n} \sum_{i=1}^n X_i - \mu \right| \geq \epsilon \right) = 0
$$

これは、サンプル平均 $\frac{1}{n} \sum_{i=1}^n X_i$ が期待値 $\mu$ に収束することを示しています。

### 結論
大数の弱法則は、確率変数の平均がその期待値に近づくことを示す強力なツールであり、確率論や統計学において重要な役割を果たします。この法則を利用することで、少数のデータからでも全体の傾向を推測することが可能になります。

```python
import numpy as np
import matplotlib.pyplot as plt

# サイコロの目の期待値
true_mean = 3.5

# サイコロを振る回数
num_rolls = 10000

# サイコロを振るシミュレーション
rolls = np.random.randint(1, 7, size=num_rolls)

# サンプル平均を計算
sample_means = np.cumsum(rolls) / np.arange(1, num_rolls + 1)

# プロット
plt.figure(figsize=(10, 6))
plt.axhline(y=true_mean, color='r', linestyle='-', label='True Mean (3.5)')
plt.plot(sample_means, label='Sample Mean')
plt.xlabel('Number of Rolls')
plt.ylabel('Mean')
plt.legend()
plt.title('Law of Large Numbers Simulation')
plt.show()
```

## 分布収束

分布収束（Convergence in Distribution）とは、確率論における確率変数の収束概念の一つであり、ある確率変数列が他の確率変数に近づく様子を記述します。これは主に確率変数列の分布が、ある特定の確率分布に収束することを意味します。

### 定義
確率変数列 $\{X_n\}$ が確率変数 $X$ に分布収束するとは、任意の実数 $x$ に対して次が成り立つときに言います。

$$
\lim_{n \to \infty} F_{X_n}(x) = F_X(x)
$$

ここで、$F_{X_n}(x)$ および $F_X(x)$ は、それぞれ $X_n$ および $X$ の累積分布関数（CDF）です。この収束は、$X_n$ の分布が $X$ の分布に近づくことを意味します。

### 具体例
以下に、標準正規分布に収束する例を示します。中心極限定理（Central Limit Theorem）により、独立同分布の確率変数の平均は、元の分布がどのようなものであっても、サンプルサイズが大きくなるにつれて標準正規分布に従うことが示されています。

Pythonでのシミュレーションコードを以下に示します。



このコードは、母集団からサンプルを取り、その平均を標準化したものが標準正規分布に収束する様子を示しています。

### 結論
分布収束は、確率変数列がある特定の分布に近づくことを示す概念で、中心極限定理のような強力な結果を含みます。この概念は、確率論や統計学において非常に重要であり、特に大規模なデータ解析において役立ちます。

```python
import numpy as np
import matplotlib.pyplot as plt

# 母集団のサイズ
n = 1000

# サンプルサイズ
sample_size = 100

# 元の分布（例: 一様分布）
population = np.random.uniform(0, 1, n)

# 標本平均の計算
sample_means = [np.mean(np.random.choice(population, sample_size)) for _ in range(n)]

# 標準化
standardized_sample_means = (sample_means - np.mean(population)) / (np.std(population) / np.sqrt(sample_size))

# プロット
plt.figure(figsize=(10, 6))
plt.hist(standardized_sample_means, bins=30, density=True, alpha=0.6, color='g', label='Sample Means')
x = np.linspace(-4, 4, 1000)
plt.plot(x, (1/np.sqrt(2 * np.pi)) * np.exp(-x**2 / 2), label='Standard Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.title('Convergence in Distribution Example')
plt.show()
```

## 小数法則

小数法則（Law of Small Numbers）は、大数の法則に対する対照的な概念として理解されることが多いですが、特定の確率論や統計の文脈で異なる意味を持ちます。一般的には、稀に発生する事象が小規模なサンプルサイズでも見かけられることや、Poisson分布のような確率分布に関する話題を指します。

### Poisson分布における小数法則

Poisson分布は、単位時間や単位空間における稀な事象の発生回数をモデル化する分布です。特に、独立した稀な事象の発生数を記述するのに適しています。以下に、Poisson分布に関連する小数法則の具体例を示します。

#### Poisson分布の定義
Poisson分布は、次の確率質量関数（PMF）によって定義されます：

$$
P(X = k) = \frac{\lambda^k e^{-\lambda}}{k!}
$$

ここで、$X$ は事象の発生回数、$\lambda$ は単位時間や単位空間における事象の平均発生率、$k$ は発生回数です。

### 具体例

### 結論
小数法則は、特にPoisson分布に関連して稀な事象の発生をモデル化する場合に使用される概念です。これは大数の法則とは異なり、小規模なサンプルサイズでも有意な情報を提供するため、特に統計学や確率論において重要です。このような稀な事象の発生を理解することで、現実世界の様々な問題に対処するための有用な洞察が得られます。

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import factorial

# Poisson分布のパラメータ
lambda_value = 3

# サンプル数
num_samples = 1000

# Poisson分布に従う乱数を生成
samples = np.random.poisson(lambda_value, num_samples)

# プロット
plt.figure(figsize=(10, 6))
plt.hist(samples, bins=range(0, max(samples) + 1), density=True, alpha=0.6, color='b', edgecolor='black')

# 理論的なPoisson分布をプロット
x = np.arange(0, max(samples) + 1)
poisson_pmf = (np.exp(-lambda_value) * lambda_value**x) / factorial(x)
plt.plot(x, poisson_pmf, 'ro', label='Poisson PMF')

plt.xlabel('Number of Floods')
plt.ylabel('Probability')
plt.legend()
plt.title('Poisson Distribution Example')
plt.show()
```

## 極値分布

極値分布（Extreme Value Distribution）とは、データセットの極端な値（最大値や最小値）を記述するための確率分布です。これは、特に大規模なデータセットにおける異常値や極端な事象の発生確率を評価する際に重要です。

### 極値理論
極値理論は、独立同分布の確率変数列における最大値や最小値の漸近分布を研究する理論です。代表的な極値分布には以下の3種類があります：

1. **Gumbel分布**（Type I極値分布）
2. **Fréchet分布**（Type II極値分布）
3. **Weibull分布**（Type III極値分布）

これらは合わせて一般化された極値分布（Generalized Extreme Value Distribution, GEV）と呼ばれます。

### Gumbel分布
Gumbel分布は、特に最大値の分布をモデル化する際に使用されます。累積分布関数（CDF）は次のように表されます：

$$
F(x) = \exp\left(-\exp\left(-\frac{x - \mu}{\beta}\right)\right)
$$

ここで、$\mu$ は位置パラメータ、$\beta$ は尺度パラメータです。

### Fréchet分布
Fréchet分布は、特に上限が存在しない場合の極値の分布をモデル化する際に使用されます。累積分布関数は次のように表されます：

$$
F(x) = \begin{cases}
0 & \text{if } x \leq 0 \\
\exp\left(-x^{-\alpha}\right) & \text{if } x > 0
\end{cases}
$$

ここで、$\alpha$ は形状パラメータです。

### Weibull分布
Weibull分布は、特に上限が存在する場合の極値の分布をモデル化する際に使用されます。累積分布関数は次のように表されます：

$$
F(x) = \begin{cases}
\exp\left(-(x/\lambda)^\kappa\right) & \text{if } x \geq 0 \\
0 & \text{if } x < 0
\end{cases}
$$

ここで、$\lambda$ は尺度パラメータ、$\kappa$ は形状パラメータです。

### 極値分布のシミュレーション
以下は、Pythonを用いたGumbel分布のシミュレーション例です。

### 結論
極値分布は、データセットの中の極端な値を理解し、モデル化するための強力なツールです。特に、自然災害のリスク評価や金融リスク管理など、極端な事象の影響を評価するために広く使用されます。極値理論を用いることで、これらの稀なイベントの発生確率をより正確に予測することが可能となります。

```python

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gumbel_r

# Gumbel分布のパラメータ
mu = 0
beta = 1

# サンプル数
num_samples = 1000

# Gumbel分布に従う乱数を生成
samples = gumbel_r.rvs(loc=mu, scale=beta, size=num_samples)

# ヒストグラムをプロット
plt.figure(figsize=(10, 6))
plt.hist(samples, bins=30, density=True, alpha=0.6, color='g', edgecolor='black', label='Sample Data')

# 理論的なGumbel分布のPDFをプロット
x = np.linspace(min(samples), max(samples), 1000)
pdf = gumbel_r.pdf(x, loc=mu, scale=beta)
plt.plot(x, pdf, 'r-', label='Gumbel PDF')

plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.title('Gumbel Distribution Example')
plt.show()
```

## 連続写像定理

連続写像定理（Continuous Mapping Theorem）は、確率論において極限分布を扱う際に非常に重要な結果です。この定理は、収束する確率変数の関数も収束することを保証します。特に、確率変数の列がある確率変数に分布収束する場合、その連続関数も同じ極限に分布収束することを示します。

### 定理の内容
確率変数の列 $\{X_n\}$ が確率変数 $X$ に分布収束し、$g$ が連続関数であるとします。このとき、$g(X_n)$ は $g(X)$ に分布収束します。つまり、

$$
X_n \xrightarrow{d} X \quad \Rightarrow \quad g(X_n) \xrightarrow{d} g(X)
$$

ここで、$X_n \xrightarrow{d} X$ は $X_n$ が $X$ に分布収束することを意味します。

### 直感的な説明
連続写像定理の直感的な理解としては、確率変数 $X_n$ が $X$ に近づくとき、連続関数 $g$ を適用した結果も $g(X)$ に近づくというものです。これは、連続関数の性質に依存しており、不連続な関数では必ずしも成立しない点に注意が必要です。

### 具体例
以下に具体的な例を示します。

#### 例1: 標準正規分布への収束
中心極限定理により、独立同分布の確率変数の標本平均は標準正規分布に収束します。この結果を利用して、連続写像定理を適用します。


この例では、標準正規分布に従うデータの標本平均を計算し、その平方根をプロットしています。連続写像定理により、標本平均が標準正規分布に収束するため、その平方根も適切な分布に収束します。

### 結論
連続写像定理は、確率変数の収束に関する非常に強力なツールであり、確率論や統計学におけるさまざまな問題に適用できます。この定理により、連続関数を適用した確率変数の列も同様に収束することが保証されるため、極限分布の解析や確率変数の変換において重要な役割を果たします。

```python
import numpy as np
import matplotlib.pyplot as plt

# サンプル数
n = 1000

# 正規分布に従うデータを生成
data = np.random.normal(loc=0, scale=1, size=(n, 10))

# 標本平均を計算
sample_means = np.mean(data, axis=1)

# 標本平均の平方根を計算 (連続関数 g(x) = sqrt(x) を適用)
sqrt_sample_means = np.sqrt(np.abs(sample_means))

# プロット
plt.figure(figsize=(12, 6))

# 標本平均のヒストグラム
plt.subplot(1, 2, 1)
plt.hist(sample_means, bins=30, density=True, alpha=0.6, color='g', edgecolor='black')
plt.title('Sample Means Distribution')

# 標本平均の平方根のヒストグラム
plt.subplot(1, 2, 2)
plt.hist(sqrt_sample_means, bins=30, density=True, alpha=0.6, color='b', edgecolor='black')
plt.title('Square Root of Sample Means Distribution')

plt.tight_layout()
plt.show()
```

## 連続写像定理とカイ二乗分布の関係

連続写像定理とカイ二乗分布の関係について説明します。連続写像定理は、確率変数の収束性に関する一般的な結果であり、カイ二乗分布は確率変数の二乗和の分布です。これらの関係を理解するために、連続写像定理の適用例としてカイ二乗分布の性質を利用します。

### カイ二乗分布の定義

カイ二乗分布（Chi-Squared Distribution）は、独立した標準正規分布に従う確率変数の二乗和の分布です。自由度 $k$ のカイ二乗分布は、次のように定義されます。

$$
Y = \sum_{i=1}^{k} Z_i^2
$$

ここで、$Z_i$ は独立な標準正規分布 $N(0, 1)$ に従う確率変数です。

### 連続写像定理の適用例

連続写像定理をカイ二乗分布に適用する例として、標準正規分布に従う確率変数の二乗和の漸近分布を考えます。

#### 例: 標準正規分布からカイ二乗分布への収束

1. $X_i \sim N(0, 1)$ が独立であるとします。
2. $T_n = \frac{1}{\sqrt{n}} \sum_{i=1}^n X_i^2$ の分布を考えます。
3. 中心極限定理により、$\frac{1}{\sqrt{n}} \sum_{i=1}^n (X_i^2 - 1)$ は正規分布に収束します。

具体的には、

$$
\frac{1}{\sqrt{n}} \sum_{i=1}^n (X_i^2 - 1) \xrightarrow{d} N(0, 2)
$$

これは、$X_i^2 - 1$ の標準偏差が $\sqrt{2}$ であることを利用しています。

4. ここで、関数 $g(x) = x + 1$ を適用すると、

$$
T_n = \frac{1}{\sqrt{n}} \sum_{i=1}^n X_i^2 = \frac{1}{\sqrt{n}} \sum_{i=1}^n (X_i^2 - 1) + \sqrt{n}
$$

したがって、

$$
T_n \xrightarrow{d} N(\sqrt{n}, 2)
$$


## デルタ法

デルタ法（Delta Method）は、推定量の関数の漸近的な分布を導くための手法です。これは、推定量が大標本のときに正規分布に従うという特性を利用し、その関数もまた漸近的に正規分布に従うことを示します。デルタ法は、主に統計学において推定値の分布やその標準誤差を求める際に使用されます。

### 基本概念

デルタ法は、以下のように定式化されます。

1. 確率変数 $X_n$ が $X$ に対して次の条件を満たすとします。
   $$
   \sqrt{n}(X_n - \theta) \xrightarrow{d} N(0, \sigma^2)
   $$

2. $g$ が $\theta$ の周りで微分可能な関数であるとします。このとき、関数 $g(X_n)$ の漸近分布は次のように得られます。
   $$
   \sqrt{n}(g(X_n) - g(\theta)) \xrightarrow{d} N(0, [g'(\theta)]^2 \sigma^2)
   $$

ここで、$g'(\theta)$ は $g$ の $\theta$ における一階微分です。

### 直感的な理解

デルタ法の基本的なアイデアは、関数 $g$ のテイラー展開を利用して、$X_n$ の変動が $g(X_n)$ にどのように影響するかを解析することです。具体的には、$X_n$ の小さな変動が $g(X_n)$ に与える影響を一次の微分 $g'(\theta)$ で近似します。

### 具体例

#### 例1: 標本平均の関数

標本平均 $\bar{X}$ が母平均 $\mu$ に対して次のように正規分布に収束する場合を考えます。
$$
\sqrt{n}(\bar{X} - \mu) \xrightarrow{d} N(0, \sigma^2)
$$

関数 $g(x) = x^2$ を考えた場合、$\bar{X}^2$ の漸近分布を求めます。

1. $g'(\mu) = 2\mu$
2. デルタ法を適用すると、次のようになります。
   $$
   \sqrt{n}(\bar{X}^2 - \mu^2) \xrightarrow{d} N(0, 4\mu^2 \sigma^2)
   $$

### 結論

デルタ法は、推定量の関数の漸近分布を求めるための有力な手法です。特に、大標本理論に基づく統計解析において、その有用性は非常に高いです。この方法を用いることで、複雑な関数の漸近分布も比較的容易に解析できるようになります。

```python

import numpy as np
import matplotlib.pyplot as plt

# 母平均と母分散
mu = 2
sigma = 1

# 標本数
n = 1000
num_samples = 10000

# 標本平均を生成
samples = np.random.normal(loc=mu, scale=sigma, size=(num_samples, n))
sample_means = np.mean(samples, axis=1)

# 関数 g(x) = x^2 を適用
g_sample_means = sample_means**2

# 理論的な漸近分布のパラメータ
asymptotic_mean = mu**2
asymptotic_std = np.sqrt((4 * mu**2 * sigma**2) / n)

# プロット
plt.figure(figsize=(10, 6))
plt.hist(g_sample_means, bins=30, density=True, alpha=0.6, color='g', edgecolor='black', label='Sample Data')

# 理論的な正規分布をプロット
x = np.linspace(min(g_sample_means), max(g_sample_means), 1000)
plt.plot(x, (1/(asymptotic_std * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - asymptotic_mean) / asymptotic_std)**2), 'r-', label='Asymptotic Normal Distribution')

plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.title('Delta Method Example')
plt.show()
```
