## 歪度

歪度は「平均値まわりの3次のモーメント」を$\sigma^3$で割ったものです。確率変数$X$の平均値を$\mu$、分散を$\sigma^2$とすると、歪度は次の式から求めることができます。

$$
\text{skewness} = \frac{E[(X - \mu)^3]}{\sigma^3}
$$

分布が左右対称の場合、$E[(X - \mu)^3] = 0$となります。したがって、歪度＝0となります。

「右裾が長い」分布の場合、平均より大きく正に離れた値があるため$E[(X - \mu)^3] > 0$となります。したがって、歪度は正の値を取ります。

「左裾が長い」分布の場合、平均より大きく負に離れた値があるため$E[(X - \mu)^3] < 0$となります。したがって、歪度は負の値を取ります。

分布が左右対称の場合に歪度=0となることを、正規分布を使って確認してみます。正規分布 $N(\mu, \sigma^2)$ のモーメント母関数は次のようになります。

$$
M_X(t) = e^{\mu t + \frac{\sigma^2 t^2}{2}}
$$

ここでは計算を簡単にするため、$\mu = 0$の正規分布 $N(0, \sigma^2)$ を考えます。上の式から、$\mu = 0$の正規分布 $N(0, \sigma^2)$ のモーメント母関数は

$$
M_X(t) = e^{\frac{\sigma^2 t^2}{2}}
$$

となります。この式をtで1回微分すると

$$
M'_X(t) = \sigma^2 t e^{\frac{\sigma^2 t^2}{2}}
$$

となります。tで2回微分すると

$$
M''_X(t) = \sigma^2 \left( e^{\frac{\sigma^2 t^2}{2}} + \sigma^2 t^2 e^{\frac{\sigma^2 t^2}{2}} \right)
$$

となります。tで3回微分すると

$$
M'''_X(t) = \sigma^4 \left( 3t e^{\frac{\sigma^2 t^2}{2}} + \sigma^2 t^3 e^{\frac{\sigma^2 t^2}{2}} \right)
$$

となります。この式に$t = 0$を代入すると、次の式が導出されます。

$$
M'''_X(0) = E[X^3] = 0
$$

したがって、$\mu = 0$の正規分布 $N(0, \sigma^2)$ の歪度は0になります。

$$
\text{skewness} = \frac{E[(X - \mu)^3]}{\sigma^3} = \frac{E[(X - 0)^3]}{\sigma^3} = \frac{E[X^3]}{\sigma^3} = \frac{0}{\sigma^3} = 0
$$

## 尖度

尖度は「平均値まわりの4次のモーメント」を$\sigma^4$で割ったものです。確率変数$X$の平均値を$\mu$、分散を$\sigma^2$とすると、尖度は次の式から求めることができます。

$$
\text{kurtosis} = \frac{E[(X - \mu)^4]}{\sigma^4}
$$

尖度は必ず0以上になります。正規分布の場合（最小値）は尖度＝3となり、3より大きいかどうかで分布の裾の重さを評価します。そのため、次の式を使って、正規分布の尖度＝0を基準とする場合の尖度を計算する場合があります。

$$
\text{kurtosis} = \frac{E[(X - \mu)^4]}{\sigma^4} - 3
$$

「裾が重い」分布の場合、平均$\mu$より大きく離れた値が多いことから、$\sigma$はより大きな値を取ります。したがって、正規分布の尖度＝3を基準とする場合は尖度はより小さな値を取り、正規分布の尖度＝0を基準とする場合は尖度は負の値を取ります。

「裾が軽い」分布の場合、平均$\mu$より大きく離れた値が少ないことから、$\sigma$はより小さな値を取ります。したがって、正規分布の尖度＝3を基準とする場合は尖度はより大きな値を取り、正規分布の尖度＝0を基準とする場合は尖度は正の値を取ります。

## 偏相関係数について

偏相関係数（partial correlation coefficient）は、3つ以上の変数の関係において、1つの変数の影響を取り除いた上での2つの変数間の相関を測定する指標です。偏相関係数は、ある変数を制御した状態での2つの変数間の純粋な相関を示します。

例えば、3つの変数 $X$, $Y$, $Z$ がある場合、$Z$ の影響を取り除いた $X$ と $Y$ の相関係数を求めることができます。

### 偏相関係数の定義

$X$ と $Y$ の間の偏相関係数を $r_{XY \cdot Z}$ と表し、以下のように計算されます：

$$
r_{XY \cdot Z} = \frac{r_{XY} - r_{XZ} r_{YZ}}{\sqrt{(1 - r_{XZ}^2)(1 - r_{YZ}^2)}}
$$

ここで：
- $r_{XY}$ は $X$ と $Y$ の相関係数
- $r_{XZ}$ は $X$ と $Z$ の相関係数
- $r_{YZ}$ は $Y$ と $Z$ の相関係数

### 標準偏差 $\sigma$ を使った偏相関係数の表現

偏相関係数を標準偏差 $\sigma$ を使って表すには、以下のようにします。

まず、3つの変数 $X$, $Y$, $Z$ の共分散行列を考えます。共分散行列 $\Sigma$ は以下のように表されます：

$$
\Sigma = \begin{pmatrix}
\sigma_{XX} & \sigma_{XY} & \sigma_{XZ} \\
\sigma_{YX} & \sigma_{YY} & \sigma_{YZ} \\
\sigma_{ZX} & \sigma_{ZY} & \sigma_{ZZ}
\end{pmatrix}
$$

ここで、$\sigma_{XX}$ は $X$ の分散、$\sigma_{XY}$ は $X$ と $Y$ の共分散などを表します。

相関係数は共分散を標準偏差で正規化したものであるため、以下の関係が成り立ちます：

$$
r_{XY} = \frac{\sigma_{XY}}{\sigma_X \sigma_Y}
$$

同様に、

$$
r_{XZ} = \frac{\sigma_{XZ}}{\sigma_X \sigma_Z}, \quad r_{YZ} = \frac{\sigma_{YZ}}{\sigma_Y \sigma_Z}
$$

これらの相関係数を偏相関係数の式に代入すると、偏相関係数を標準偏差を使って表すことができます。

### 偏相関係数の計算例

具体的に、標準偏差を使って偏相関係数を計算する例を示します。

例えば、以下のような標準偏差と共分散が与えられている場合を考えます：

- $\sigma_X = 2$, $\sigma_Y = 3$, $\sigma_Z = 1$
- $\sigma_{XY} = 4$, $\sigma_{XZ} = 2$, $\sigma_{YZ} = 1$

これらの値を使って相関係数を計算します：

$$
r_{XY} = \frac{\sigma_{XY}}{\sigma_X \sigma_Y} = \frac{4}{2 \times 3} = \frac{4}{6} = \frac{2}{3}
$$

$$
r_{XZ} = \frac{\sigma_{XZ}}{\sigma_X \sigma_Z} = \frac{2}{2 \times 1} = \frac{2}{2} = 1
$$

$$
r_{YZ} = \frac{\sigma_{YZ}}{\sigma_Y \sigma_Z} = \frac{1}{3 \times 1} = \frac{1}{3}
$$

次に、これらの相関係数を使って偏相関係数を計算します：

$$
r_{XY \cdot Z} = \frac{r_{XY} - r_{XZ} r_{YZ}}{\sqrt{(1 - r_{XZ}^2)(1 - r_{YZ}^2)}} = \frac{\frac{2}{3} - 1 \times \frac{1}{3}}{\sqrt{(1 - 1^2)(1 - \left(\frac{1}{3}\right)^2)}} = \frac{\frac{2}{3} - \frac{1}{3}}{\sqrt{(0)(1 - \frac{1}{9})}} = \frac{\frac{1}{3}}{0} = 0
$$

この結果は、$r_{XZ}$ が 1 であるため、偏相関係数が0であることを示しています。これは、$Z$ が $X$ と $Y$ の間の関係を完全に説明しているため、$Z$ の影響を取り除くと $X$ と $Y$ の間に相関がなくなることを示しています。

このように、偏相関係数を用いることで、ある変数の影響を制御した上で他の変数間の純粋な関係を評価することができます。

## 条件付き期待値の定義、意味

確率変数 $Y$ の値が $y$ であるという条件のもとでの $X$ の期待値を $E[X | Y = y]$ と書き、条件付き期待値と言います。「$Y = y$ となるグループに限定したときの $X$ の平均」とも言えます。

式で書くと（離散型確率変数の場合）、

$$
E[X | Y = y] = \sum_x x \frac{P(X = x, Y = y)}{P(Y = y)}
$$

となります。

定義から分かるように、**条件付き期待値 $E[X | Y = y]$ は $y$ の関数です。**

$E[X | Y = y] = f(y)$ のことを $E[X | Y] = f(Y)$ と書くことも多いです。また、期待値を取る確率変数が $X$ であることを強調するために $E_X[X | Y]$ と書くこともあります。

また、条件付き分散 $V[X | Y]$ も同様に定義されます（グループ内での分散、$Y$ の関数）。条件付きでない場合と同様に、以下が成立します：

$$
V[X | Y] = E[X^2 | Y] - (E[X | Y])^2
$$

（というよりこの式を条件付き分散の定義とすることが多い気がします）
