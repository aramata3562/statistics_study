## プロポーショナル・ハザードモデル（比例ハザードモデル）について

プロポーショナル・ハザードモデル（比例ハザードモデル、Cox比例ハザードモデルとも呼ばれる）は、生存時間分析に広く使用される統計モデルです。これは、特定の時間までに事象（例えば、死亡、故障、離脱など）が発生するリスクを解析するための手法です。

### 基本概念

比例ハザードモデルは、以下のようにハザード関数を定義します：

$$
h(t | \mathbf{x}) = h_0(t) \exp(\beta_1 x_1 + \beta_2 x_2 + \cdots + \beta_p x_p)
$$

ここで、
- $h(t | \mathbf{x})$ は、時間 $t$ におけるハザード関数です。これは、時間 $t$ において事象が発生する即時リスクを表します。
- $h_0(t)$ は、基準ハザード関数（ベースラインハザード関数）です。これは、すべての説明変数がゼロの場合のハザード率です。
- $\beta_1, \beta_2, \ldots, \beta_p$ は回帰係数です。
- $x_1, x_2, \ldots, x_p$ は説明変数（共変量）です。

### モデルの仮定

比例ハザードモデルの主な仮定は、説明変数の効果が時間に依存せず比例的であることです。具体的には、説明変数の効果がハザード関数に対して時間と無関係に一定の比率で影響を与えると仮定します。

### パラメータ推定

比例ハザードモデルのパラメータ（$\beta$）は、部分尤度（partial likelihood）を最大化することで推定されます。この方法は、ベースラインハザード関数を明示的に推定することなく、説明変数の効果を評価できます。

### 生存分析の用語

- **生存時間（Survival Time）**:
    観測開始から事象（例えば、死亡、故障、離脱など）が発生するまでの時間です。

- **検閲（Censoring）**:
    事象が発生せずに観測が終了することです。例えば、研究期間が終了するまでに死亡しなかった患者は右側検閲（右検閲）されます。

- **ハザード関数（Hazard Function）**:
    ある時点で事象が発生する即時リスクを表します。

### Pythonコード例

以下に、Pythonを用いた比例ハザードモデルの実装例を示します。ここでは、`lifelines` パッケージを使用してモデルを適用します。


### 結果の解釈

モデルを適用すると、以下のような情報が得られます：

- **係数（$\beta$）**:
    各説明変数に対する回帰係数が表示されます。正の係数は、変数が増加するにつれてハザードが増加することを示し、負の係数はハザードが減少することを示します。

- **p値**:
    係数の統計的有意性を評価するためのp値が表示されます。p値が小さいほど、その変数がモデルにとって重要であることを示します。

- **ハザード比（Hazard Ratio）**:
    係数の指数関数を取った値であり、変数が1単位増加したときのハザード比率を示します。ハザード比が1より大きければリスクが増加し、1より小さければリスクが減少します。

例えば、以下のような出力が得られるとします：

```
                   coef  exp(coef)  se(coef)    z      p  lower 0.95  upper 0.95
fin             -0.379    0.685    0.191  -1.988  0.047     -0.754      -0.004
age             -0.057    0.944    0.022  -2.575  0.010     -0.100      -0.014

```

ここで、`fin` 変数の係数が -0.379 である場合、ハザード比は $exp(-0.379) = 0.685$ です。これは、`fin` が1単位増加すると、ハザードが0.685倍になること、すなわちリスクが減少することを示します。

比例ハザードモデルは、生存時間データの解析において非常に強力なツールであり、様々な分野で広く応用されています。

````python
import pandas as pd
from lifelines import CoxPHFitter
from lifelines.datasets import load_rossi

# データセットの読み込み
data = load_rossi()

# データフレームの内容を表示
print(data.head())

# Cox比例ハザードモデルの作成
cph = CoxPHFitter()
cph.fit(data, duration_col='week', event_col='arrest')

# モデルの要約を表示
cph.print_summary()

# 回帰係数のプロット
cph.plot()
````

## カプラン・マイヤー推定量について

カプラン・マイヤー推定量（Kaplan-Meier estimator）は、生存分析において生存関数（Survival Function）を推定するための非パラメトリック手法です。これは、時間とともにあるイベント（例えば、死亡、故障、離脱など）が発生する確率を推定します。カプラン・マイヤー推定量は、検閲データ（Censored Data）を考慮して計算されます。

### 生存関数とは

生存関数 $S(t)$ は、時間 $t$ までにイベントが発生しない確率を表します。これは次のように定義されます：

$$
S(t) = P(T > t)
$$

ここで、$T$ はイベントが発生する時間です。

### カプラン・マイヤー推定量の計算

カプラン・マイヤー推定量は、観測されたイベントタイムポイントに基づいて生存関数を推定します。具体的には、各タイムポイントでの生存確率を次のように計算します：

1. すべてのイベントタイムポイントを順序付けてリストします。
2. 各タイムポイント $t_i$ での生存確率 $S(t_i)$ を以下のように計算します：

$$
S(t_i) = S(t_{i-1}) \left( 1 - \frac{d_i}{n_i} \right)
$$

ここで、
- $S(t_{i-1})$ は前のタイムポイントまでの生存確率です。
- $d_i$ はタイムポイント $t_i$ でイベントが発生した個体数です。
- $n_i$ はタイムポイント $t_i$ まで生存していた個体数です。

### 検閲データ

検閲データは、イベントが発生せずに観察が終了するデータを指します。カプラン・マイヤー推定量は、検閲データを適切に処理するため、イベントが発生しなかった場合でも生存確率の推定に影響を与えません。

### カプラン・マイヤー曲線

カプラン・マイヤー推定量を用いて生存関数をプロットしたものがカプラン・マイヤー曲線です。この曲線は、時間の経過とともに生存確率がどのように変化するかを視覚的に示します。

### Pythonコード例

以下に、Pythonを用いたカプラン・マイヤー推定量の計算とカプラン・マイヤー曲線のプロットの例を示します。ここでは `lifelines` ライブラリを使用します。



### 結果の解釈

カプラン・マイヤー曲線は、時間の経過とともに生存確率がどのように変化するかを示します。この曲線を分析することで、特定のタイムポイントまでの生存確率や、異なるグループ間の生存率の比較ができます。

例えば、複数の治療群の生存曲線を比較することで、どの治療法がより効果的であるかを評価することができます。カプラン・マイヤー推定量は、臨床試験や信頼性工学など、様々な分野で広く応用されています。

### 例：複数グループの比較

複数のグループ（例えば、異なる治療法を受けた患者グループ）の生存曲線を比較する場合、以下のようにしてカプラン・マイヤー推定量を計算し、プロットします。



このコードでは、グループAとグループBの生存曲線を比較しています。異なるグループ間で生存確率の違いを視覚的に確認することができます。

````python
import pandas as pd
from lifelines import KaplanMeierFitter
import matplotlib.pyplot as plt

# サンプルデータの作成
data = {
    'time': [5, 6, 6, 6, 7, 10, 13, 16, 22, 23, 6, 9, 10, 11, 17, 20, 25],
    'event': [1, 0, 1, 1, 0, 1, 1, 0, 1, 0, 1, 0, 1, 1, 0, 1, 0]
}
df = pd.DataFrame(data)

# カプラン・マイヤー推定量のフィッティング
kmf = KaplanMeierFitter()
kmf.fit(durations=df['time'], event_observed=df['event'])

# 生存曲線のプロット
kmf.plot_survival_function()
plt.title('Kaplan-Meier Survival Curve')
plt.xlabel('Time')
plt.ylabel('Survival Probability')
plt.show()
````

````python
import pandas as pd
from lifelines import KaplanMeierFitter
import matplotlib.pyplot as plt

# サンプルデータの作成（異なるグループの例）
data = {
    'time': [5, 6, 6, 6, 7, 10, 13, 16, 22, 23, 6, 9, 10, 11, 17, 20, 25],
    'event': [1, 0, 1, 1, 0, 1, 1, 0, 1, 0, 1, 0, 1, 1, 0, 1, 0],
    'group': ['A', 'A', 'A', 'A', 'A', 'A', 'A', 'A', 'A', 'A', 'B', 'B', 'B', 'B', 'B', 'B', 'B']
}
df = pd.DataFrame(data)

# グループごとにカプラン・マイヤー推定量を計算
kmf = KaplanMeierFitter()

plt.figure()

for name, grouped_df in df.groupby('group'):
    kmf.fit(durations=grouped_df['time'], event_observed=grouped_df['event'], label=name)
    kmf.plot_survival_function()

plt.title('Kaplan-Meier Survival Curves by Group')
plt.xlabel('Time')
plt.ylabel('Survival Probability')
plt.show()
````
