## 期待値と分散の計算

このページでは、確率変数の期待値（平均値）と分散を定義し、その計算方法を説明しています。以下に、その内容を詳細に説明します。

### 確率変数と期待値

確率変数 $X$ が取る値を $x$ とし、その確率を $P(X = x)$ または $p(x)$ とします。サイコロの例で考えると、$X$ は1から6の値を取り、各値の確率 $p(x)$ は $\frac{1}{6}$ です。

期待値 $\mathbb{E}[X]$（または平均値 $\mu$）は、次の式で定義されます。

$$
\mu = \mathbb{E}[X] = \sum_{x} x p(x)
$$

ここで、$\sum_{x}$ は $X$ が取りうる全ての値に関する和を表します。

### 一般的な関数の期待値

確率変数 $X$ の関数 $g(X)$ の期待値は次のように定義されます。

$$
\mathbb{E}[g(X)] = \sum_{x} g(x)p(x)
$$

特に、$g(x) = (x - \mu)^2$ を偏差の2乗として、分散 $\text{Var}(X)$ は次のように定義されます。

### 分散の計算

分散 $\sigma^2$（または $\text{Var}(X)$）は、期待値からの偏差の2乗の期待値として定義されます。

$$
\sigma^2 = \text{Var}(X) = \mathbb{E}[(X - \mu)^2] = \sum_{x} (x - \mu)^2 p(x)
$$

ここで、$(X - \mu)^2$ を展開してから $p(x)$ を掛けて和を取ると、次のように表せます。

$$
(X - \mu)^2 = X^2 - 2X\mu + \mu^2
$$

これを期待値の形式に当てはめます。

$$
\text{Var}(X) = \mathbb{E}[X^2 - 2X\mu + \mu^2]
$$

期待値の線形性を利用すると、

$$
\text{Var}(X) = \mathbb{E}[X^2] - 2\mu\mathbb{E}[X] + \mu^2
$$

ここで、$\mathbb{E}[X] = \mu$ であることから、

$$
\text{Var}(X) = \mathbb{E}[X^2] - 2\mu^2 + \mu^2
$$

つまり、

$$
\text{Var}(X) = \mathbb{E}[X^2] - \mu^2
$$

これにより、確率変数 $X$ の分散 $\text{Var}(X)$ は、$X^2$ の期待値から $X$ の期待値の2乗を引いたものとなります。

以上が、期待値と分散の計算手順の詳細です。

## 分散の計算について

分散の計算手順についてもう少し具体的な例を使って説明します。これにより、理解をさらに深められると思います。

### 具体例：サイコロの分散

ここでは、サイコロを振ったときの目の分散を計算してみます。サイコロの目の値は $X = 1, 2, 3, 4, 5, 6$ であり、各目が出る確率は $p(x) = \frac{1}{6}$ です。

1. **期待値の計算**：

まず、期待値 $\mu = \mathbb{E}[X]$ を計算します。

$$
\mu = \mathbb{E}[X] = \sum_{x=1}^{6} x \cdot p(x) = \sum_{x=1}^{6} x \cdot \frac{1}{6}
$$

具体的に計算すると、

$$
\mu = \frac{1}{6}(1 + 2 + 3 + 4 + 5 + 6) = \frac{1}{6} \cdot 21 = 3.5
$$

2. **二乗期待値の計算**：

次に、$X^2$ の期待値 $\mathbb{E}[X^2]$ を計算します。

$$
\mathbb{E}[X^2] = \sum_{x=1}^{6} x^2 \cdot p(x) = \sum_{x=1}^{6} x^2 \cdot \frac{1}{6}
$$

具体的に計算すると、

$$
\mathbb{E}[X^2] = \frac{1}{6}(1^2 + 2^2 + 3^2 + 4^2 + 5^2 + 6^2) = \frac{1}{6}(1 + 4 + 9 + 16 + 25 + 36) = \frac{1}{6} \cdot 91 = \frac{91}{6}
$$

3. **分散の計算**：

分散 $\sigma^2 = \text{Var}(X)$ を求めます。

$$
\sigma^2 = \mathbb{E}[X^2] - \mu^2
$$

ここで、

$$
\mathbb{E}[X^2] = \frac{91}{6}, \quad \mu = 3.5
$$

したがって、

$$
\sigma^2 = \frac{91}{6} - (3.5)^2
$$

$$
(3.5)^2 = 12.25 

$$
$$
\sigma^2 = \frac{91}{6} - 12.25 = \frac{91}{6} - \frac{73.5}{6} = \frac{91 - 73.5}{6} = \frac{17.5}{6} = 2.9167
$$

これで、サイコロを振ったときの目の分散が2.9167であることがわかりました。

以上が具体的な例を使った期待値と分散の計算手順です。このようにして、分布の基本的な統計量を求めることができます。

## 連続確率変数の期待値と分散

このページでは、連続的な確率変数の期待値と分散について説明しています。離散的な確率変数と異なり、連続的な確率変数では確率密度関数（PDF）を用いて期待値や分散を計算します。

### 確率密度関数

連続確率変数 $X$ の確率密度関数（PDF） $f(x)$ は次のように定義されます。

$$
f(x) = \lim_{\epsilon \to 0} \frac{P(x < X \leq x + \epsilon)}{\epsilon}
$$

これは、区間 $[x, x + \epsilon]$ における $X$ の確率を $\epsilon$ で割って、$\epsilon$ が0に近づく極限を取ることで、$x$ における確率密度を求めるものです。

### 期待値の計算

連続確率変数 $X$ の期待値 $\mathbb{E}[X]$ は次のように定義されます。

$$
\mathbb{E}[X] = \int_{-\infty}^{\infty} x f(x) \, dx
$$

この積分は、$X$ の全ての値に対する $x$ の加重平均を表します。

### 分散の計算

連続確率変数 $X$ の分散 $\text{Var}(X)$ は次のように定義されます。

$$
\text{Var}(X) = \int_{-\infty}^{\infty} (x - \mu)^2 f(x) \, dx
$$

ここで、$\mu$ は期待値 $\mathbb{E}[X]$ です。

### 例題

ここに記載されている例題を元に、具体的な計算の流れを説明します。

#### 例題 1.1

**問題 1**:
男女共学のA大学の入学試験では、ある年、受験生に占める女子の比率は0.4で、合格率は女子が0.5、男子が0.4であった。この大学全体の合格率を求めよ。

**解答**:
全体の合格率は、女子と男子の合格率の加重平均として計算できます。まず、女子と男子のそれぞれの合格率を求めます。

女子の割合 $= 0.4$
男子の割合 $= 0.6$ （全体から女子の割合を引いたもの）

女子の合格率 $= 0.5$
男子の合格率 $= 0.4$

全体の合格率 $\mu$ は次のように計算します。

$$
\mu = 0.4 \cdot 0.5 + 0.6 \cdot 0.4 = 0.2 + 0.24 = 0.44
$$

したがって、この大学全体の合格率は 44% です。

**問題 2**:
この大学の合格者からランダムに選んだ1人が女子である確率を求めよ。

**解答**:
ベイズの定理を用いて、この確率を計算します。まず、合格者全体に占める女子の割合を求めます。

全体の合格者の数に対する女子の合格者の割合を計算します。

女子の合格者の割合 $= 0.4 \cdot 0.5 = 0.2$
男子の合格者の割合 $= 0.6 \cdot 0.4 = 0.24$

合格者全体の中で女子の割合を求めます。

$$
P(\text{女子 | 合格}) = \frac{0.2}{0.2 + 0.24} = \frac{0.2}{0.44} = \frac{5}{11}
$$

したがって、合格者の中からランダムに選んだ1人が女子である確率は $\frac{5}{11}$ です。

以上が連続確率変数の期待値と分散、ならびに例題の解説です。期待値や分散の計算方法を理解することで、様々な確率問題を解く際の基礎が身につきます。
