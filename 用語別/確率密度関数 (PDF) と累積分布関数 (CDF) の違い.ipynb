## 確率密度関数 (PDF) と累積分布関数 (CDF) の違い

確率密度関数 (Probability Density Function, PDF) と累積分布関数 (Cumulative Distribution Function, CDF) は、確率分布を記述するために使用される重要な概念です。それぞれの違いとその役割を説明します。

### 確率密度関数 (PDF)

確率密度関数は、連続確率変数の値の分布を記述する関数です。具体的には、ある範囲内にある値が出現する確率を示します。PDFは値そのものの確率を示すわけではなく、確率の密度を示します。

- **定義**: 確率変数 $X$ の確率密度関数 $f_X(x)$ は、次の条件を満たします。
  1. $f_X(x) \geq 0 \quad \forall x$
  2. $\int_{-\infty}^{\infty} f_X(x) \, dx = 1$

- **意味**: $a$ から $b$ の範囲に $X$ が存在する確率は、次の積分で表されます。
  $$
  P(a \leq X \leq b) = \int_{a}^{b} f_X(x) \, dx
  $$

### 累積分布関数 (CDF)

累積分布関数は、ある値以下の確率を示す関数です。これは、確率変数が特定の値以下である確率を累積的に計算します。

- **定義**: 確率変数 $X$ の累積分布関数 $F_X(x)$ は次のように定義されます。
  $$
  F_X(x) = P(X \leq x) = \int_{-\infty}^{x} f_X(t) \, dt
  $$

- **意味**: $X$ が値 $x$ 以下である確率を示します。

### PDF と CDF の関係

PDF と CDF は密接な関係があります。CDF は PDF の積分であり、PDF は CDF の微分です。

- **CDF から PDF**: 累積分布関数 $F_X(x)$ が与えられたとき、その確率密度関数 $f_X(x)$ は次のように得られます。
  $$
  f_X(x) = \frac{d}{dx} F_X(x)
  $$

- **PDF から CDF**: 確率密度関数 $f_X(x)$ が与えられたとき、その累積分布関数 $F_X(x)$ は次のように得られます。
  $$
  F_X(x) = \int_{-\infty}^{x} f_X(t) \, dt
  $$

### 具体例
標準正規分布を例にとって説明します。

- **標準正規分布の PDF**:
  $$
  f_X(x) = \frac{1}{\sqrt{2\pi}} e^{-\frac{x^2}{2}}
  $$

- **標準正規分布の CDF**:
  $$
  F_X(x) = \int_{-\infty}^{x} \frac{1}{\sqrt{2\pi}} e^{-\frac{t^2}{2}} \, dt
  $$

### Python を用いた具体例




このコードでは、標準正規分布のPDFとCDFをプロットしています。PDFは確率密度を示し、CDFは累積確率を示します。

### 結論
PDFとCDFは、確率分布を理解するための基本的なツールであり、それぞれ異なる情報を提供します。PDFは特定の値の密度を示し、CDFはある値以下の確率を示します。これらを使い分けることで、データの性質や確率の分布を詳細に理解することが可能となります。

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# 標準正規分布のパラメータ
mu = 0
sigma = 1

# xの範囲
x = np.linspace(-5, 5, 1000)

# PDF と CDF を計算
pdf = norm.pdf(x, mu, sigma)
cdf = norm.cdf(x, mu, sigma)

# プロット
plt.figure(figsize=(12, 6))

# PDF のプロット
plt.subplot(1, 2, 1)
plt.plot(x, pdf, label='PDF', color='blue')
plt.title('Probability Density Function')
plt.xlabel('x')
plt.ylabel('Density')
plt.legend()

# CDF のプロット
plt.subplot(1, 2, 2)
plt.plot(x, cdf, label='CDF', color='green')
plt.title('Cumulative Distribution Function')
plt.xlabel('x')
plt.ylabel('Cumulative Probability')
plt.legend()

plt.tight_layout()
plt.show()
```
