## モーメント母関数について

モーメント母関数（Moment Generating Function, MGF）は、確率分布のモーメント（期待値、分散、歪度など）を生成するための関数です。確率変数の特徴を解析する上で重要な役割を果たします。

### 定義
確率変数 $X$ のモーメント母関数 $M_X(t)$ は、次のように定義されます。

$$
M_X(t) = \mathbb{E}[e^{tX}]
$$

ここで、$\mathbb{E}$ は期待値を表します。この関数は、$t$ の適切な範囲内で収束することが必要です。

### 性質
モーメント母関数の重要な性質の一つは、そのテイラー展開がモーメントを含むことです。具体的には、$t = 0$ のまわりで展開すると次のようになります。

$$
M_X(t) = 1 + \mathbb{E}[X]t + \frac{\mathbb{E}[X^2]}{2!}t^2 + \frac{\mathbb{E}[X^3]}{3!}t^3 + \cdots
$$

この展開からわかるように、モーメント母関数の $t$ に関する $n$ 次導関数を $t=0$ で評価すると、確率変数の $n$ 次モーメントが得られます。

$$
\frac{d^n M_X(t)}{dt^n}\bigg|_{t=0} = \mathbb{E}[X^n]
$$

### 例：正規分布
正規分布 $N(\mu, \sigma^2)$ のモーメント母関数を計算してみます。正規分布の確率変数 $X$ のモーメント母関数は次のようになります。

$$
M_X(t) = \mathbb{E}[e^{tX}] = \exp\left(\mu t + \frac{\sigma^2 t^2}{2}\right)
$$

### 利用例
モーメント母関数は、以下のような場面で利用されます。
1. **モーメントの計算**: 分布のモーメント（平均、分散など）を直接計算する。
2. **分布の同定**: 異なる確率分布を区別するためのツールとして利用する。
3. **大数の法則や中心極限定理の証明**: 確率論の重要な定理の証明に役立つ。

モーメント母関数は、確率変数の持つ性質を解析し、理解するための強力なツールです。具体的な例を通して、その応用範囲や計算方法を学ぶことで、統計学や確率論の深い理解に繋がります。

## 正規分布のモーメント計算

正規分布 $N(\mu, \sigma^2)$ の確率変数 $X$ に対して、そのモーメント母関数を用いてモーメントを計算する過程を示します。

### モーメント母関数の導出

正規分布 $X \sim N(\mu, \sigma^2)$ の場合、モーメント母関数 $M_X(t)$ は次のように表されます。

$$
M_X(t) = \mathbb{E}[e^{tX}]
$$

正規分布の確率密度関数 $f_X(x)$ は次の通りです。

$$
f_X(x) = \frac{1}{\sqrt{2\pi\sigma^2}} \exp\left(-\frac{(x - \mu)^2}{2\sigma^2}\right)
$$

モーメント母関数は以下のように計算されます。

$$
M_X(t) = \int_{-\infty}^{\infty} e^{tx} f_X(x) \, dx
$$

確率密度関数 $f_X(x)$ を代入します。

$$
M_X(t) = \int_{-\infty}^{\infty} e^{tx} \cdot \frac{1}{\sqrt{2\pi\sigma^2}} \exp\left(-\frac{(x - \mu)^2}{2\sigma^2}\right) \, dx
$$

指数関数の指数部分を整理します。

$$
e^{tx} \cdot \exp\left(-\frac{(x - \mu)^2}{2\sigma^2}\right) = \exp\left(tx - \frac{(x - \mu)^2}{2\sigma^2}\right)
$$

指数の中身を一つにまとめます。

$$
tx - \frac{(x - \mu)^2}{2\sigma^2} = tx - \frac{x^2 - 2\mu x + \mu^2}{2\sigma^2}
= -\frac{x^2 - 2\sigma^2tx + 2\mu x - \mu^2}{2\sigma^2}
$$

これを更に整理します。

$$
= -\frac{x^2 - 2(\mu + \sigma^2 t)x + \mu^2}{2\sigma^2}
= -\frac{x^2 - 2(\mu + \sigma^2 t)x + (\mu + \sigma^2 t)^2 - (\mu + \sigma^2 t)^2 + \mu^2}{2\sigma^2}
$$

$$
= -\frac{(x - (\mu + \sigma^2 t))^2}{2\sigma^2} + \frac{(\mu + \sigma^2 t)^2 - \mu^2}{2\sigma^2}
$$

この結果を用いてモーメント母関数を再度書き直します。

$$
M_X(t) = \int_{-\infty}^{\infty} \exp\left(-\frac{(x - (\mu + \sigma^2 t))^2}{2\sigma^2}\right) \exp\left(\frac{(\mu + \sigma^2 t)^2 - \mu^2}{2\sigma^2}\right) \frac{1}{\sqrt{2\pi\sigma^2}} \, dx
$$

最初の部分は正規分布の確率密度関数の形になっています。

$$
M_X(t) = \exp\left(\frac{(\mu + \sigma^2 t)^2 - \mu^2}{2\sigma^2}\right)
$$

これを整理します。

$$
= \exp\left(\frac{\mu^2 + 2\mu\sigma^2 t + \sigma^4 t^2 - \mu^2}{2\sigma^2}\right)
= \exp\left(\frac{2\mu\sigma^2 t + \sigma^4 t^2}{2\sigma^2}\right)
= \exp\left(\mu t + \frac{\sigma^2 t^2}{2}\right)
$$

### モーメントの計算

モーメント母関数が得られたので、モーメントを計算するためにその $t$ に関する導関数を $t=0$ で評価します。

1. **平均（一次モーメント）**：

$$
M_X'(t) = \frac{d}{dt} \left[ \exp\left(\mu t + \frac{\sigma^2 t^2}{2}\right) \right]
= \exp\left(\mu t + \frac{\sigma^2 t^2}{2}\right) \left( \mu + \sigma^2 t \right)
$$

$$
M_X'(0) = \left. \exp\left(\mu t + \frac{\sigma^2 t^2}{2}\right) \left( \mu + \sigma^2 t \right) \right|_{t=0} = \mu
$$

2. **分散（二次モーメント）**：

二次モーメントを計算するために二次導関数を求めます。

$$
M_X''(t) = \frac{d^2}{dt^2} \left[ \exp\left(\mu t + \frac{\sigma^2 t^2}{2}\right) \right]
$$

まず、一次導関数を再度確認します。

$$
M_X'(t) = \exp\left(\mu t + \frac{\sigma^2 t^2}{2}\right) \left( \mu + \sigma^2 t \right)
$$

これをもう一度微分します。

$$
M_X''(t) = \frac{d}{dt} \left[ \exp\left(\mu t + \frac{\sigma^2 t^2}{2}\right) \left( \mu + \sigma^2 t \right) \right]
$$

積の微分法則を適用します。

$$
M_X''(t) = \exp\left(\mu t + \frac{\sigma^2 t^2}{2}\right) \left( \mu + \sigma^2 t \right)^2 + \exp\left(\mu t + \frac{\sigma^2 t^2}{2}\right) \sigma^2
$$

$$
M_X''(0) = \left. \exp\left(\mu t + \frac{\sigma^2 t^2}{2}\right) \left( \mu + \sigma^2 t \right)^2 + \exp\left(\mu t + \frac{\sigma^2 t^2}{2}\right) \sigma^2 \right|_{t=0}
$$

$$
= \exp(0) \left( \mu^2 \right) + \exp(0) \sigma^2
= \mu^2 + \sigma^2
$$

二次モーメント $\mathbb{E}[X^2]$ は $\mu^2 + \sigma^2$ ですが、分散 $\text{Var}(X)$ は以下のように計算されます。

$$
\text{Var}(X) = \mathbb{E}[X^2] - (\mathbb{E}[X])^2 = (\mu^2 + \sigma^2) - \mu^2 = \sigma^2
$$
このようにして、正規分布の一次モーメントと二次モーメントをモーメント母関数から計算することができます。
