## ミンコフスキー距離について

ミンコフスキー距離（Minkowski Distance）は、ベクトル空間における距離を測るための一般的な距離計量法の一つです。ミンコフスキー距離は、ユークリッド距離やマンハッタン距離など、他の多くの距離計量法の一般化された形です。距離の測定に使用されるパラメータ $p$ によって異なる種類の距離を表現できます。

### ミンコフスキー距離の定義

2つの点 $x = (x_1, x_2, \ldots, x_n)$ と $y = (y_1, y_2, \ldots, y_n)$ に対して、ミンコフスキー距離は次のように定義されます：

$$
D_p(x, y) = \left( \sum_{i=1}^n |x_i - y_i|^p \right)^{\frac{1}{p}}
$$

ここで、$p$ はミンコフスキー距離のパラメータであり、$p \geq 1$ です。

### 特殊な場合

- **マンハッタン距離（p = 1）**：
  $$
  D_1(x, y) = \sum_{i=1}^n |x_i - y_i|
  $$
  マンハッタン距離は、各次元の絶対差の合計です。市街距離とも呼ばれ、碁盤の目のような道のりを想像するとわかりやすいです。

- **ユークリッド距離（p = 2）**：
  $$
  D_2(x, y) = \sqrt{\sum_{i=1}^n (x_i - y_i)^2}
  $$
  ユークリッド距離は、2点間の直線距離です。

- **チェビシェフ距離（p → ∞）**：
  $$
  D_\infty(x, y) = \max_{i} |x_i - y_i|
  $$
  チェビシェフ距離は、各次元の差の中で最大のものを距離とします。

### ミンコフスキー距離の特性

1. **対称性**：$D_p(x, y) = D_p(y, x)$
2. **非負性**：$D_p(x, y) \geq 0$
3. **自己同一性**：$D_p(x, y) = 0 \iff x = y$
4. **三角不等式**：$D_p(x, z) \leq D_p(x, y) + D_p(y, z)$

### まとめ

ミンコフスキー距離は、パラメータ $p$ に応じて様々な距離計量を表現できる柔軟な距離尺度です。ユークリッド距離やマンハッタン距離などの特殊な場合を包含しており、さまざまなデータ分析や機械学習の応用において利用されています。パラメータ $p$ を調整することで、データの性質に応じた最適な距離計量を選択することが可能です。
