## 離散一様分布の期待値と分散の導出

離散一様分布は、$a$ から $b$ までの $n$ 個の整数が等確率で発生する分布です。$n$ 個の値は $a, a+1, a+2, \ldots, b$ であり、各値の確率は $P(X = k) = \frac{1}{n}$ です。ここで、$n = b - a + 1$ です。

### 期待値の導出

離散一様分布の期待値 $E[X]$ は次の式で計算されます：

$$
E[X] = \sum_{k=a}^{b} k \cdot P(X = k)
$$

各値の確率 $P(X = k) = \frac{1}{n}$ なので、

$$
E[X] = \sum_{k=a}^{b} k \cdot \frac{1}{n}
$$

これを簡略化すると、

$$
E[X] = \frac{1}{n} \sum_{k=a}^{b} k
$$

ここで、$a$ から $b$ までの整数の和は、等差数列の和の公式を使って求められます：

$$
\sum_{k=a}^{b} k = \frac{n}{2} (a + b)
$$

したがって、

$$
E[X] = \frac{1}{n} \cdot \frac{n}{2} (a + b) = \frac{a + b}{2}
$$

### 分散の導出

離散一様分布の分散 $Var(X)$ は次の式で計算されます：

$$
Var(X) = E[X^2] - (E[X])^2
$$

まず、$E[X^2]$ を計算します：

$$
E[X^2] = \sum_{k=a}^{b} k^2 \cdot P(X = k)
$$

各値の確率 $P(X = k) = \frac{1}{n}$ なので、

$$
E[X^2] = \sum_{k=a}^{b} k^2 \cdot \frac{1}{n}
$$

これを簡略化すると、

$$
E[X^2] = \frac{1}{n} \sum_{k=a}^{b} k^2
$$

ここで、$a$ から $b$ までの整数の二乗の和は次の公式で求められます：

$$
\sum_{k=a}^{b} k^2 = \frac{(b-a+1)(2b^2 + 2ab + 2a^2 + 3b + 3a + 1)}{6}
$$

したがって、

$$
E[X^2] = \frac{1}{n} \cdot \frac{(b-a+1)(2b^2 + 2ab + 2a^2 + 3b + 3a + 1)}{6}
$$

ここで $n = b - a + 1$ を代入すると、

$$
E[X^2] = \frac{2b^2 + 2ab + 2a^2 + 3b + 3a + 1}{6}
$$

次に、$E[X]$ の二乗を計算します：

$$
(E[X])^2 = \left( \frac{a + b}{2} \right)^2 = \frac{(a + b)^2}{4}
$$

したがって、

$$
Var(X) = E[X^2] - (E[X])^2 = \frac{2b^2 + 2ab + 2a^2 + 3b + 3a + 1}{6} - \frac{(a + b)^2}{4}
$$

ここで、分母を揃えるために通分します：

$$
Var(X) = \frac{4(2b^2 + 2ab + 2a^2 + 3b + 3a + 1) - 3(a^2 + 2ab + b^2)}{12}
$$

これを整理すると、

$$
Var(X) = \frac{8b^2 + 8ab + 8a^2 + 12b + 12a + 4 - 3a^2 - 6ab - 3b^2}{12}
$$

さらに簡略化すると、

$$
Var(X) = \frac{5b^2 + 2ab + 5a^2 + 12b + 12a + 4}{12}
$$

したがって、離散一様分布の分散は次のようになります：

$$
Var(X) = \frac{(b-a+1)^2 - 1}{12}
$$

## 離散一様分布の確率母関数

離散一様分布の確率母関数 (Probability Generating Function, PGF) を求めるには、まず確率母関数の定義を確認します。確率母関数 $G_X(t)$ は次のように定義されます：

$$
G_X(t) = E[t^X] = \sum_{k=a}^{b} t^k \cdot P(X = k)
$$

ここで、離散一様分布の確率は $P(X = k) = \frac{1}{n}$ （$n = b - a + 1$）です。したがって、

$$
G_X(t) = \sum_{k=a}^{b} t^k \cdot \frac{1}{n}
$$

これを簡略化すると、

$$
G_X(t) = \frac{1}{n} \sum_{k=a}^{b} t^k
$$

### 等比数列の和の公式

$\sum_{k=a}^{b} t^k$ は等比数列の和として表せます。一般的に、等比数列の和は次の公式を使って求められます：

$$
\sum_{k=0}^{n} r^k = \frac{r^{n+1} - 1}{r - 1}
$$

しかし、ここでは和の範囲が $a$ から $b$ なので、次のように調整します：

$$
\sum_{k=a}^{b} t^k = \sum_{k=0}^{b} t^k - \sum_{k=0}^{a-1} t^k
$$

これを等比数列の和の公式を使って計算すると、

$$
\sum_{k=a}^{b} t^k = \frac{t^{b+1} - 1}{t - 1} - \frac{t^a - 1}{t - 1}
$$

これをまとめると、

$$
\sum_{k=a}^{b} t^k = \frac{t^{b+1} - 1 - t^a + 1}{t - 1} = \frac{t^{b+1} - t^a}{t - 1}
$$

### 確率母関数の最終形

この結果を確率母関数の式に代入します：

$$
G_X(t) = \frac{1}{n} \cdot \frac{t^{b+1} - t^a}{t - 1}
$$

したがって、離散一様分布 $U(a, b)$ の確率母関数は次のようになります：

$$
G_X(t) = \frac{t^{b+1} - t^a}{n(t - 1)}
$$

ここで、$n = b - a + 1$ です。
