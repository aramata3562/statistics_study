### 問題 8.3 野球の打率

昨シーズンを2割8分の打率で終わった打者が、今シーズンもこの確率でヒットを打つものとし、450打数であるとすると、3割バッターになれる確率はどれくらいか。またこのバッターが、確率0.2以上で3割バッターになろうとすると、打数はどのくらいでなければならないか。

#### 問題 1: 450打数で3割バッターになる確率
打率は成功率と考え、これは二項分布に基づく確率問題です。

- 昨シーズンの打率 $p = 0.28$
- 今シーズンの打数 $n = 450$
- 求めたいのは、3割バッター（打率0.3以上）になる確率

打数に対する成功数（ヒット数）を $X$ とすると、これを二項分布 $X \sim \text{Binomial}(n, p)$ で表します。この場合、打率0.3以上は $X \geq 135$ に相当します（450打数の30%）。

##### 確率計算:
二項分布の確率を計算するためには、各ヒット数に対する確率を足し合わせます。具体的には、次のように計算します。
$$ P(X \geq 135) = \sum_{k=135}^{450} \binom{450}{k} (0.28)^k (0.72)^{450-k} $$

これは直接計算するのは難しいので、通常は正規分布に近似します。二項分布 $X$ の期待値 $\mu$ と分散 $\sigma^2$ は以下のように求められます。
$$ \mu = np = 450 \times 0.28 = 126 $$
$$ \sigma^2 = np(1-p) = 450 \times 0.28 \times 0.72 \approx 90.72 $$
$$ \sigma \approx 9.52 $$

正規分布に近似すると、標準化された変数 $Z$ に変換できます。
$$ Z = \frac{X - \mu}{\sigma} = \frac{135 - 126}{9.52} \approx 0.945 $$

標準正規分布の表を用いて、$Z = 0.945$ 以上の確率を求めると、およそ0.173になります。したがって、450打数で3割バッターになる確率は約17.3%です。

#### 問題 2: 3割バッターになるための打数
次に、3割バッターになる確率が20%以上となるための打数を求めます。これは逆に計算し、必要な打数 $n$ を求めます。

- 成功確率 $p = 0.28$
- 3割バッターになる確率 $P(X \geq 0.3n) \geq 0.2$

確率を標準正規分布に近似して考えると、
$$ Z = \frac{0.3n - 0.28n}{\sqrt{0.28 \times 0.72 \times n}} $$
$$ Z = \frac{0.02n}{\sqrt{0.2016n}} = \frac{0.02\sqrt{n}}{\sqrt{0.2016}} = \frac{0.02\sqrt{n}}{0.448} \approx 0.04464\sqrt{n} $$

これが0.2の確率以上になるための $Z$ 値を標準正規分布の表から求めると、約0.84です。
$$ 0.04464\sqrt{n} \geq 0.84 $$
$$ \sqrt{n} \geq \frac{0.84}{0.04464} \approx 18.82 $$
$$ n \geq (18.82)^2 \approx 354 $$

したがって、3割バッターになる確率が20%以上になるためには、最低でも約354打数が必要です。
