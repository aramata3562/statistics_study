## ベルヌーイ分布

ベルヌーイ分布は、確率論と統計学における最も基本的な離散確率分布の一つです。ベルヌーイ分布は、試行が成功か失敗かの二つの結果のみを持つ試行、すなわちベルヌーイ試行に関連しています。

### 定義
ベルヌーイ分布は、以下のように定義されます。試行が成功する確率を $p$ とし、失敗する確率を $1 - p$ とします。このとき、確率変数 $X$ がベルヌーイ分布に従う場合、その確率質量関数（PMF）は以下のようになります：

$$
P(X = x) = 
\begin{cases} 
p & \text{if } x = 1 \\
1 - p & \text{if } x = 0 
\end{cases}
$$

ここで、$0 \leq p \leq 1$ です。

### 特性

- **期待値（平均値）**: ベルヌーイ分布の期待値は成功確率 $p$ です。
  $$
  E[X] = p
  $$

- **分散**: ベルヌーイ分布の分散は $p(1 - p)$ です。
  $$
  \text{Var}(X) = p(1 - p)
  $$

### Pythonでの実装例

ベルヌーイ分布をPythonでシミュレーションするためには、`numpy`ライブラリを使用できます。


このコードは、成功確率 $p = 0.5$ のベルヌーイ分布に従う1000個のサンプルを生成し、その期待値と分散を計算します。

```python
import numpy as np

# 成功確率 p
p = 0.5

# サンプルサイズ n
n = 1000

# ベルヌーイ分布に従うサンプルを生成
samples = np.random.binomial(1, p, n)

# サンプルの期待値（平均値）と分散を計算
mean = np.mean(samples)
variance = np.var(samples)

print(f"サンプルの期待値（平均値）: {mean}")
print(f"サンプルの分散: {variance}")
```

### 応用例
ベルヌーイ分布は、様々な応用があります。例えば、コインの表と裏を考える場合や、製品が合格か不合格かを評価する場合などです。ベルヌーイ分布を拡張したものに二項分布があり、これは複数のベルヌーイ試行の成功回数をモデル化します。

ベルヌーイ分布は確率論の基礎であり、他の複雑な分布や統計モデルの理解にも重要です。

## ポアソン分布

ポアソン分布は、ある特定の時間区間や空間区間における特定の事象の発生回数をモデル化するために使用される確率分布です。特に、発生する事象が独立であり、平均発生率が一定である場合に適用されます。

### 定義

ポアソン分布は、平均発生率 $\lambda$ をパラメータとして持ちます。確率変数 $X$ が時間区間内に発生する事象の数を表すとき、その確率質量関数（PMF）は次のように定義されます：

$$
P(X = k) = \frac{\lambda^k e^{-\lambda}}{k!} \quad \text{for} \quad k = 0, 1, 2, \ldots
$$

ここで、
- $k$ は非負整数（発生する事象の回数）
- $\lambda$ は平均発生率（単位時間あたりの事象の平均発生回数）
- $e$ は自然対数の底（約2.71828）

### 特性

ポアソン分布の主な特性は以下の通りです：

- **期待値（平均値）**: $\lambda$
  $$
  E[X] = \lambda
  $$

- **分散**: $\lambda$
  $$
  \text{Var}(X) = \lambda
  $$

### 例

ポアソン分布は、様々な現実のシナリオに適用されます。例えば、
- 1時間あたりに受ける電話の平均本数
- 1日に郵便受けに届く手紙の平均枚数
- ある交差点で1時間に発生する交通事故の平均件数

### Pythonでの実装例

Pythonを用いてポアソン分布をシミュレーションし、グラフを描写する方法を示します。

```python
import numpy as np
import matplotlib.pyplot as plt

# 平均発生率 λ
lambda_ = 4

# ポアソン分布に従うサンプルを生成
n = 1000
samples = np.random.poisson(lambda_, n)

# サンプルの期待値（平均値）と分散を計算
mean = np.mean(samples)
variance = np.var(samples)

print(f"Mean of samples: {mean}")
print(f"Variance of samples: {variance}")

# サンプルのヒストグラムを描写
plt.hist(samples, bins=range(min(samples), max(samples) + 1), edgecolor='black', alpha=0.75)
plt.xlabel('Number of events k')
plt.ylabel('Frequency')
plt.title(f'Poisson Distribution (λ = {lambda_})')
plt.show()
```

このコードは、平均発生率 $\lambda = 4$ のポアソン分布に従う1000個のサンプルを生成し、その期待値と分散を計算して、ヒストグラムを描写します。

### まとめ

ポアソン分布は、一定の時間や空間における事象の発生回数をモデル化するために非常に有用な確率分布です。その簡潔な数式と明確な特性により、さまざまな統計的な問題に適用されています。

## ポアソン分布が予測に使える理由

ポアソン分布は、特定の時間や空間内での事象の発生回数を予測するのに非常に有用です。その理由は以下の通りです：

### 1. 事象の独立性と発生率の一定性

ポアソン分布は、次の条件を満たす事象に適用できます：

1. **事象の独立性**: 任意の2つの事象の発生は独立している。
2. **発生率の一定性**: 単位時間あたりの平均発生率（$\lambda$）が一定である。

これらの条件が満たされる場合、ポアソン分布は特定の時間内に何回事象が発生するかをモデル化するのに適しています。

### 2. 希少事象のモデリング

ポアソン分布は、特定の時間枠内での希少事象の発生をモデリングするのに特に有効です。例えば、以下のようなシナリオがあります：

- 一定期間における交通事故の発生回数
- 1時間あたりのコールセンターへの電話の件数
- 特定の病気の発症率

これらの事象は通常、発生する回数が比較的少なく、しかしその発生の期待値が一定であるため、ポアソン分布が適用されます。

### 3. 大数の法則と小区間への分割

ポアソン分布は、ある大きな時間枠を小さな区間に分割し、それぞれの区間で事象が発生するかどうかを二項分布として扱うことからも導き出せます。区間が十分に小さく、事象が稀である場合、二項分布はポアソン分布に近似されます。これにより、ポアソン分布は以下の条件を満たす際に自然に出現します：

- 小区間における発生確率が非常に小さい
- 全体の発生率が一定である

### 4. 実用的な適用例

ポアソン分布の実用例は多岐にわたります。以下にいくつかの例を示します：

- **テレコム**: ネットワークトラフィックのピーク時のパケット到着率を予測する。
- **金融**: 一定期間内の特定の金融イベント（例：デフォルト、詐欺）の発生回数を予測する。
- **医療**: 病院の緊急治療室における1日あたりの患者到着数を予測する。

### 数学的根拠

ポアソン分布の数式は次の通りです：

$$
P(X = k) = \frac{\lambda^k e^{-\lambda}}{k!}
$$

この式により、単位時間内に事象がちょうど $k$ 回発生する確率が計算できます。このように、事象の発生回数を具体的に予測できるため、ポアソン分布は予測において非常に強力なツールとなります。

### まとめ

ポアソン分布が予測に使える理由は、その数学的な性質と実用的な適用性にあります。特に、事象が独立しており発生率が一定である場合、ポアソン分布はその発生回数を正確にモデル化し、予測するのに非常に適しています。
