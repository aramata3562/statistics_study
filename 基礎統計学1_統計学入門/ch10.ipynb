### 問題 10.1

#### 問題の概要:
1. 天秤で重さ100gの物体を測定する。1回ごとの測定には誤差が生じるが、その誤差分布は平均0、分散0.1の正規分布 $\ N(0, 0.1) $ であるとする。測定を10回繰り返すとき、測定値の平均 $\overline{X}$ の標本分布を求めよ。また、$\overline{X} - 100| > 0.3 $ となる確率を計算せよ。

2. 標本の大きさの決定: 上の問題と同一の天秤を使い、 $\ |\overline{X} - 100| < 0.1 $ となる確率を、0.9以上とするためには、何回測定を繰り返せばよいか。

#### 解説

##### 測定値の平均 $\ \overline{X} $ の標本分布
個々の測定値を $\ X_i $ とすると、各 $\ X_i $ は独立に $\ N(100, 0.1) $ に従います。測定値の平均 $\ \overline{X} $ は以下のように表されます。
$$ \overline{X} = \frac{1}{n} \sum_{i=1}^{n} X_i $$

ここで、$\ n = 10 $ 回の測定を行うとき、$\ \overline{X} $ の期待値 $\ E(\overline{X}) $ と分散 $\ \text{Var}(\overline{X}) $ は以下のようになります。
$$ E(\overline{X}) = 100 $$
$$ \text{Var}(\overline{X}) = \frac{0.1}{n} = \frac{0.1}{10} = 0.01 $$

したがって、$\ \overline{X} $ の標本分布は次のようになります。
$$ \overline{X} \sim N(100, 0.01) $$

##### $\ |\overline{X} - 100| > 0.3 $ となる確率
標準正規分布 $\ Z $ に変換すると、次のように求められます。
$$ Z = \frac{\overline{X} - 100}{\sqrt{0.01}} = \frac{\overline{X} - 100}{0.1} $$

したがって、$\ |\overline{X} - 100| > 0.3 $ となる確率は次のようになります。
$$ P(|\overline{X} - 100| > 0.3) = P\left(\left|Z\right| > \frac{0.3}{0.1}\right) = P(|Z| > 3) $$

標準正規分布において、$\ |Z| > 3 $ となる確率は非常に小さく、具体的には約0.0027 (0.27%) です。

### 問題 10.2

##### 標本の大きさの決定
上の問題と同一の天秤を使い、$\ |\overline{X} - 100| < 0.1 $ となる確率を0.9以上とするためには、何回測定を繰り返せばよいかを求めます。

同様に、標準正規分布 $\ Z $ に変換すると、次のように表されます。
$$ Z = \frac{\overline{X} - 100}{\sqrt{\frac{0.1}{n}}} = \frac{\overline{X} - 100}{\sqrt{0.1/n}} $$

$\ |\overline{X} - 100| < 0.1 $ となる確率を0.9以上とするためには、次のようになります。
$$ P\left(\left|\frac{\overline{X} - 100}{\sqrt{0.1/n}}\right| < 1\right) \geq 0.9 $$

標準正規分布の表を用いて、$\ P(|Z| < 1) \approx 0.6827 $ です。したがって、
$$ 0.6827 = 0.9 $$

これは逆正規分布を用いて計算します。逆正規分布を用いると、次のようになります。
$$ \frac{0.1}{\sqrt{0.1/n}} = 1.645 $$

したがって、必要な測定回数 $\ n $ は次のようになります。
$$ \sqrt{0.1/n} = \frac{0.1}{1.645} \approx 0.0608 $$
$$ n \approx \frac{0.1}{0.0608^2} \approx 27.1 $$

したがって、0.9以上の確率で $\ |\overline{X} - 100| < 0.1 $ となるためには、最低でも約28回の測定が必要です。


### 問題 10.3

#### 問題の概要:
母平均 $\mu = 4$、母分散 $\sigma^2 = 15$ の正規母集団から、大きさ $n = 10$ の標本を抽出する。

1. 標本平均 $\overline{X}$ が 3 と 6 の間にある確率を求めよ。
2. 標本分散 $S^2$ が $a$ を超える確率が 0.05 となるような定数 $a$ の値を求めよ。

#### 解説

##### 標本平均 $\overline{X}$ が 3 と 6 の間にある確率
標本平均 $\overline{X}$ の期待値 $E(\overline{X})$ と分散 $\text{Var}(\overline{X})$ は以下のようになります。
$$ E(\overline{X}) = \mu = 4 $$
$$ \text{Var}(\overline{X}) = \frac{\sigma^2}{n} = \frac{15}{10} = 1.5 $$

したがって、標本平均 $\overline{X}$ の標準偏差は $\sqrt{1.5}$ です。標準正規分布 $Z$ に変換すると、次のようになります。
$$ Z = \frac{\overline{X} - 4}{\sqrt{1.5}} $$

$\overline{X}$ が 3 と 6 の間にある確率は次のようになります。
$$ P(3 < \overline{X} < 6) = P\left(\frac{3 - 4}{\sqrt{1.5}} < Z < \frac{6 - 4}{\sqrt{1.5}}\right) $$
$$ = P\left(-\frac{1}{\sqrt{1.5}} < Z < \frac{2}{\sqrt{1.5}}\right) $$
$$ = P(-0.816 < Z < 1.633) $$

標準正規分布表を用いて、この範囲の確率を求めると、およそ0.7454になります。

##### 標本分散 $S^2$ が $a$ を超える確率
標本分散 $S^2$ は $\chi^2$ 分布に従います。具体的には、次のようになります。
$$ \frac{(n-1)S^2}{\sigma^2} \sim \chi^2_{n-1} $$

ここで、$n = 10$ なので、自由度は $9$ です。確率が $0.05$ となる $a$ を求めるために、$\chi^2$ 分布表を用います。
$$ P\left(\chi^2_{9} > \frac{9a}{15}\right) = 0.05 $$

$\chi^2_{9}$ 分布の上側5%点は約16.92です。
$$ \frac{9a}{15} = 16.92 $$
$$ a = \frac{16.92 \times 15}{9} \approx 28.2 $$

したがって、$a \approx 28.2$ となります。

### 問題 10.4

#### 問題の概要:
母平均 $\mu = 3$ の正規母集団（$\sigma^2$ は未知）から、大きさ $n = 15$ の標本を抽出する。$s$ を標本標準偏差とするとき、$(\overline{X} - 3)/s > a$ となる確率が 0.01 となるような定数 $a$ の値を求めよ。

#### 解説

標本平均 $\overline{X}$ は次のように表されます。
$$ \overline{X} \sim t_{n-1} $$

ここで、自由度 $n-1 = 14$ です。確率が $0.01$ となる $a$ を求めるために、$t$ 分布表を用います。
$$ P\left(t_{14} > a\right) = 0.01 $$

$t_{14}$ 分布の上側1%点は約2.624です。したがって、$a \approx 2.624$ となります。

### 問題 10.5

#### 問題の概要:
母平均 $\mu_1 = 2$、母分散 $\sigma_1^2 = 3$ の正規母集団から大きさ $m = 10$ の標本を抽出し、母平均 $\mu_2 = 5$、母分散 $\sigma_2^2 = 4$ の正規母集団から大きさ $n = 8$ の標本を抽出する。二つの標本平均の差の標本分布を求めよ。

#### 解説

標本平均の差 $\overline{X}_1 - \overline{X}_2$ の期待値と分散を求めます。

- $\overline{X}_1$ の期待値 $E(\overline{X}_1) = \mu_1 = 2$
- $\overline{X}_1$ の分散 $\text{Var}(\overline{X}_1) = \frac{\sigma_1^2}{m} = \frac{3}{10} = 0.3$
- $\overline{X}_2$ の期待値 $E(\overline{X}_2) = \mu_2 = 5$
- $\overline{X}_2$ の分散 $\text{Var}(\overline{X}_2) = \frac{\sigma_2^2}{n} = \frac{4}{8} = 0.5$

標本平均の差の期待値と分散は次のようになります。
$$ E(\overline{X}_1 - \overline{X}_2) = E(\overline{X}_1) - E(\overline{X}_2) = 2 - 5 = -3 $$
$$ \text{Var}(\overline{X}_1 - \overline{X}_2) = \text{Var}(\overline{X}_1) + \text{Var}(\overline{X}_2) = 0.3 + 0.5 = 0.8 $$

したがって、標本平均の差の標本分布は次のようになります。
$$ \overline{X}_1 - \overline{X}_2 \sim N(-3, 0.8) $$

### 問題 10.6

#### 問題の概要:
10.5 の二つの標本から標本分散の比 $s_1^2 / s_2^2$ を計算するとき、その比が $c$ を超える確率が 0.05 となるような定数 $c$ の値を求めよ。

#### 解説

標本分散の比 $s_1^2 / s_2^2$ は、自由度 $m-1$ と $n-1$ を持つ $F$ 分布に従います。

$$ \frac{s_1^2 / \sigma_1^2}{s_2^2 / \sigma_2^2} \sim F_{m-1, n-1} $$

ここで、$m = 10$ なので、自由度は $9$、$n = 8$ なので、自由度は $7$ です。確率が $0.05$ となる $c$ を求めるために、$F$ 分布表を用います。
$$ P\left(F_{9, 7} > c\right) = 0.05 $$

$F_{9, 7}$ 分布の上側5%点は約3.68です。したがって、$c \approx 3.68$ となります。

### 問題 10.7

#### 問題の概要:
一つまたは二つの正規母集団からの標本について、次の確率を求めよ。

1. 大きさ $n = 10$ の標本で標本平均 $\overline{X}$ が母平均 $\mu$ から $\overline{X}$ の標本標準偏差の 0.8 倍以上ずれる確率
2. 同じく、標本平均 $\overline{X}$ が母平均から $\overline{X}$ の標本標準偏差の 0.8 倍以上ずれる確率
3. 同じく、標本分散 $s^2$ が母分散の 2 倍を越えまたは 1/2 倍を下回る確率
4. 大きさ $m = n = 10$ の標本について、母平均は等しいのに、二つの標本平均が標本標準偏差（ただし2標本を併合して計算する）の3倍以上ずれる。ただし、母分散は等しいとする。
5. 同じく、二つの標本分散が 3 倍以上異なる確率

#### 解説

##### 1. 標本平均 $\overline{X}$ が母平均 $\mu$ から $\overline{X}$ の標本標準偏差の 0.8 倍以上ずれる確率

標準正規分布 $Z$ に変換すると、次のように求められます。
$$ Z = \frac{\overline{X} - \mu}{\sigma / \sqrt{n}} $$
ここで、$n = 10$、$\sigma$ は母標準偏差です。

$\overline{X}$ の標本標準偏差は $\sigma / \sqrt{n}$ なので、
$$ Z = \frac{\overline{X} - \mu}{\sigma / \sqrt{n}} \approx 0.8 $$

標準正規分布表を用いて、$Z$ が 0.8 倍以上ずれる確率を求めます。
$$ P(|Z| > 0.8) \approx 2 \times (1 - \Phi(0.8)) \approx 2 \times 0.2119 = 0.4238 $$

##### 2. 標本分散 $s^2$ が母分散の 2 倍を越えまたは 1/2 倍を下回る確率

標本分散 $s^2$ は $\chi^2$ 分布に従います。
$$ \frac{(n-1)s^2}{\sigma^2} \sim \chi^2_{n-1} $$

$n = 10$ のとき、自由度は $9$ です。$\chi^2_{9}$ 分布の上下5%点を用います。
$$ P\left(\frac{s^2}{\sigma^2} > 2\right) \approx P\left(\chi^2_{9} > 16.92\right) = 0.05 $$

したがって、標本分散 $s^2$ が母分散の2倍を越える確率は0.05です。同様に、1/2倍を下回る確率も0.05です。合わせて0.10です。

### 問題 10.8

#### 問題の概要:
身長と体重の母集団分布は2次元正規分布である。母集団における相関係数（母相関係数）が $\rho = 0.6$ であるとき、大きさ $n = 15$ の標本からの標本相関係数 $r$ に対し、$c_1 < r < c_2$ となる確率が0.95となるような定数 $c_1, c_2$ の値を求めよ。

#### 解説

標本相関係数 $r$ の分布は、自由度 $n-2$ の $t$ 分布を用いて次のように表されます。
$$ t = \frac{r \sqrt{n-2}}{\sqrt{1-r^2}} $$

ここで、$n = 15$ なので、自由度は $13$ です。$t$ 分布の上側2.5%点を求め、逆変換します。

パーセント点について次の関係式を確認せよ。

3. 自由度 $k \geq 120$ では、$t$ 分布は標準正規分布に近似できる。
