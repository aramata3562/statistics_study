# ベイズの定理
ベイズの定理は、統計学や確率論における基本的な定理であり、事後確率（ある事象が観測された後の確率）を求めるための方法を提供します。ベイズの定理は、条件付き確率の概念に基づいています。

### ベイズの定理の定義

ベイズの定理は次のように表されます：
$$
P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)}
$$

ここで、
- $P(A|B)$ は、事象 $B$ が起こったときの事象 $A$ の条件付き確率（事後確率）。
- $P(B|A)$ は、事象 $A$ が起こったときの事象 $B$ の条件付き確率（尤度）。
- $P(A)$ は、事象 $A$ が起こる確率（事前確率）。
- $P(B)$ は、事象 $B$ が起こる確率（周辺確率）。

### ベイズの定理の直感的な理解

ベイズの定理は、以下のような質問に答えるために使われます：「ある観測結果 $B$ が得られたとき、仮説 $A$ の正しさの確率はどれくらいか？」

### 具体例

例として、ある病気の検査について考えてみましょう。

#### 条件

1. 病気にかかっている確率（事前確率） $P(D) = 0.01$ （1%）。
2. 検査が病気を正しく検出する確率（真陽性率、尤度） $P(T|D) = 0.99$ （99%）。
3. 病気でない人が誤って陽性と診断される確率（偽陽性率） $P(T|\neg D) = 0.05$ （5%）。
4. 全体で陽性と診断される確率 $P(T)$ は、全ての陽性結果の合計で求められます。

#### 事後確率の計算

ベイズの定理を使用して、検査が陽性となったときに実際に病気である確率 $P(D|T)$ を求めます。

まず、全体で陽性と診断される確率 $P(T)$ を計算します。
$$
P(T) = P(T|D)P(D) + P(T|\neg D)P(\neg D)
$$
ここで、$P(\neg D) = 1 - P(D) = 0.99$ です。

したがって、
$$
P(T) = (0.99 \times 0.01) + (0.05 \times 0.99) = 0.0099 + 0.0495 = 0.0594
$$

次に、ベイズの定理を適用して $P(D|T)$ を求めます。
$$
P(D|T) = \frac{P(T|D) \cdot P(D)}{P(T)} = \frac{0.99 \times 0.01}{0.0594} \approx 0.166
$$

つまり、検査が陽性となったときに実際に病気である確率は約16.6%です。

### まとめ

ベイズの定理は、観測結果に基づいて仮説の確率を更新する強力なツールです。日常的な問題から科学的な研究まで、さまざまな分野で広く使われています。具体的な例を通じて、ベイズの定理がどのように機能するかを理解することが重要です。
